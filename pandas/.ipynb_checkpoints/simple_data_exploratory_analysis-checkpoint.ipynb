There are two types of datasets were are now going to explore.
### Beers:

- Name: Name of the beer.

```python
import pandas as pd
import seaborn as sns
import pandas_profiling
sns.set(color_codes=True)
sns.set_palette(sns.color_palette('muted'))
```

```python
beers = pd.DataFrame.from_csv("craft-beers-dataset/data/processed/beers.csv")
breweries = pd.DataFrame.from_csv("craft-beers-dataset/data/processed/breweries.csv")
beers_and_breweries = pd.merge(beers, 
                               breweries, 
                               how='inner', 
                               left_on="brewery_id", 
                               right_on="id", 
                               sort=True,
                               suffixes=('_beer', '_brewery'))
```

```python
beers.dtypes
```

```python
def get_var_category(series):
    unique_count = series.nunique(dropna=False)
    total_count = len(series)
    if pd.api.types.is_numeric_dtype(series):
        return 'Numerical'
    elif pd.api.types.is_datetime64_any_dtype(series):
        return 'Date'
    elif unique_count==total_count:
        return 'Text (Unique)'
    else:
        return 'Categorical'
```

```python
def print_categories(df):
    for column_name in df:
        print(column_name,":",get_var_category(df[column_name]))
```

```python
print_categories(beers)
```

Correlations
Correlations are a great way to discover relationships between numerical variables. The Pearson correlation coefficient is widely used that measures the linear dependence between two variables.
The correlation coefficient ranges from 1 to -1. A correlation of 1 is a total positive correlation, a correlation of -1 is a total negative correlation.

```python
beers[['ibu','abv','ounces']].corr()
```

Here correlation between ibu and itself is 1, as numerical variables are perfectly correlated with themselves.
Also, there is a 0.67 correlation between IBU and ABV. Although it is not a total positive correlation, it is an important factor to notice.
